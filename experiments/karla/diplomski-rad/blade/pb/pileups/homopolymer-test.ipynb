```python
import numpy as np
import os
from progressbar import ProgressBar
```

```python
pileup_paths = [
    './e-coli-NCTC86-all-contigs-indel-classes/pileups-X-0-indels.npy',
    './k-pneumoniae-NCTC204-all-contigs-indel-classes/pileups-X-0-indels.npy',
    './m-morgani-NCTC235-all-contigs-indel-classes/pileups-X-0-indels.npy',
    './s-enterica-NCTC92-all-contigs-indel-classes/pileups-X-0-indels.npy',
    './s-enterica-NCTC129-all-contigs-indel-classes/pileups-X-0-indels.npy'
]
labels_paths = [
    './e-coli-NCTC86-all-contigs-indel-classes/pileups-y-0.npy',
    './k-pneumoniae-NCTC204-all-contigs-indel-classes/pileups-y-0.npy',
    './m-morgani-NCTC235-all-contigs-indel-classes/pileups-y-0.npy',
    './s-enterica-NCTC92-all-contigs-indel-classes/pileups-y-0.npy',
    './s-enterica-NCTC129-all-contigs-indel-classes/pileups-y-0.npy'
]
```

```python
def convert_label_to_name(label):
    mapping = {0: 'A', 1: 'C', 2: 'G', 3: 'T', 4: 'I', 5: 'D'}
    return mapping[np.argmax(label)]
```

## E. coli

```python
pileup_path, labels_path = pileup_paths[0], labels_paths[0]


pileups = np.load(pileup_path)
labels = np.load(labels_path)
print(pileups.shape)
print(labels.shape)

print (pileups[0], labels[0])

print('A:', np.sum(labels[:, 0]))
print('C:', np.sum(labels[:, 1]))
print('G:', np.sum(labels[:, 2]))
print('T', np.sum(labels[:, 3]))
print('I:', np.sum(labels[:, 4]))
print('D:', np.sum(labels[:, 5]))

deletion_indices = np.where(np.all(labels == [0, 0, 0, 0, 0, 1, 0], axis=1))[0]

neighbourhood_size = 20

for deletion_index in deletion_indices:
    classes = [convert_label_to_name(label) for label in labels[
        deletion_index-neighbourhood_size:deletion_index+neighbourhood_size+1]]
    classes = ''.join(classes)
    classes = classes[:neighbourhood_size] + '-D-' + classes[neighbourhood_size+1:]

    print(classes)
```
